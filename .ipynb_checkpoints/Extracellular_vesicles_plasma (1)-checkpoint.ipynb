```python
%%R
library(dplyr)
library(tidyr)
library(ggplot2)
library(forcats)
```

```python
%%R
fig <- function(width, heigth){
     options(repr.plot.width = width, repr.plot.height = heigth)
}
```

```python
%%R
options(repr.plot.width = 20, repr.plot.height =10, repr.plot.res = 200)
X15_cluster_pathway_category%>% ggplot(aes(x = fct_inorder( Comparison), y = fct_inorder(Term), color = pvalue)) + 
        geom_point(aes(size = Fold_Enrichment)) +  scale_color_gradient(low="blue", high="red", guide = guide_colorbar(title = "p-value")) +   
        guides(size = guide_legend(title = "Fold Enrichment")) + labs(x = "Clusters") + 
        labs(y = " Pathways") + facet_wrap(~category, scales = "free_y", ncol = 1, strip.position = "top")
```
